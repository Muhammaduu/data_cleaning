Data cleaning is import aspect in ML and it involves number techniques, i will discuss a number of techniques in this project that will help us to clean our data to do some ML algorithms

##### our task here is prepare a calling list that contains;
##### phone numbers that are arranged in the format 000-000-0000
#### drop the unwanted columns
##### filling NAs, NANs or N/as
#### format Yes as Y and No as N
#### arrange the address columns into streer address and zip-code
#### a calling list that customers that said to be called




```python
# loading needed library
import pandas as pd
```

```python
df=pd.read_csv("Customer Call List.csv") # loading dataset
```

```python
df
```

dropping unwanted columns

```python
unwanted_columns=["Not_Useful_Column","Unnamed: 8","Unnamed: 9"]
df=df.drop(columns=unwanted_columns)
```

```python
df.head()
```

```python
df.dtypes
```

duplicates 

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
df
```

strip function

```python
df['Last_Name']=df['Last_Name'].str.strip("/")
df['Last_Name']=df['Last_Name'].str.strip("...")
df['Last_Name']=df['Last_Name'].str.strip("_")
df
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]','', regex=True)
df['Phone_Number']=df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number'].apply(lambda x: x[0:3] + '-'+ x[3:6] + '-' + x[6:10])
df['Phone_Number']=df['Phone_Number'].apply(lambda x: x[0:3] + '-'+ x[3:6] + '-' + x[6:10])
df['Phone_Number']=df['Phone_Number'].str.replace("Na--", "")
df['Phone_Number']=df['Phone_Number'].str.replace("nan--", "")
df
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('/',"")
```

```python
df
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('|',"")
```

```python
df
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('--',"-")
```

```python
df
```

```python
df[['Street_address','State','Zip_code']]=df['Address'].str.split(',', n=2, expand=True)
```

```python
df
```

```python
df['Paying Customer']=df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer']=df['Paying Customer'].str.replace('No','N')
df
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No','N')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')
df
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('nan',' ')
df
```

```python
df=df.fillna("")
df
```

```python
df=df.replace('N/a','')
df
```

```python
for i in df.index:
    if df.loc[i,'Do_Not_Contact']=='Y':
        df.drop(i, inplace=True)
```

```python
df
```

```python
for i in df.index:
    if df.loc[i,'Phone_Number']=='--':
        df.drop(i, inplace=True)
```
